```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
encoded_df = pd.get_dummies(df, columns=['Embarked'])
```

```python
true_map={True:1,False:0}
```

```python
encoded_df['Embarked_C']=encoded_df['Embarked_C'].map(true_map)
```

```python
encoded_df['Embarked_S']=encoded_df['Embarked_S'].map(true_map)
```

```python
encoded_df['Embarked_Q']=encoded_df['Embarked_Q'].map(true_map)
```

```python
df = encoded_df
```

```python
df
```

```python
gender_map={'male':0,'female':1}
```

```python
df['Sex']=df['Sex'].map(gender_map)
```

```python
df
```

```python
X = df[['Pclass', 'Sex','Age','SibSp','Fare','Embarked_C','Embarked_Q','Embarked_S']]
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.2, random_state=0)
print('Total no. of samples: Training and Testing dataset separately!')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow as tf

X_train = tf.constant(X_train)
y_train = tf.constant(y_train)
X_test = tf.constant(X_test)
y_test = tf.constant(y_test)
```

```python
import keras 
```

```python
# Create the model
model = tf.keras.Sequential([
                  tf.keras.layers.Dense(128, activation="relu"),
                  tf.keras.layers.Dense(256, activation="relu"),
                  tf.keras.layers.Dense(128, activation="relu"),
                  tf.keras.layers.Dense(1, activation="sigmoid")
 ])
# Compile the model
model.compile(loss="binary_crossentropy",
              optimizer="Adam",
              metrics=["accuracy"])

```

```python
# Fit the model
history = model.fit(X_train,
                    y_train,
                    epochs=50,
                    validation_data=(X_test, y_test),
                    verbose=0)
model.evaluate(X_test, y_test)
```

```python
y_pred_prob = model.predict(X_test)

# Convert probabilities to binary values (0 or 1)
y_pred = (y_pred_prob > 0.5).astype("int32")
```

```python
y_pred
```

```python
from sklearn import metrics 
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
print('Confusion matrix')
print(cnf_matrix)
print('---------------')
print('Precision:', metrics.precision_score(y_test, y_pred))
print('Recall:', metrics.recall_score(y_test, y_pred))
print('F1 Score:', metrics.f1_score(y_test, y_pred))
print('Accuracy Score:', metrics.accuracy_score(y_test, y_pred))
```
