```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Cabin_t']=df['Cabin'].str[0]
```

```python
df=df.drop(columns='Cabin_t')
```

```python
df
```

```python
df.info()
```

```python
encoded_df = pd.get_dummies(df, columns=['Embarked'])
```

```python
encoded_df
```

```python
true_map={True:1,False:0}
```

```python
encoded_df['Embarked_Q']=encoded_df['Embarked_Q'].map(true_map)
```

```python
encoded_df['Embarked_S']
```

```python
df = encoded_df
```

```python
gender_map={'male':0,'female':1}
```

```python
df
```

```python
df['Sex']=df['Sex'].map(gender_map)
```

```python
df
```

```python
X = df[['Pclass', 'Sex','Age','SibSp','Fare','Embarked_C','Embarked_Q','Embarked_S']]
y = df['Survived']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.2, random_state=0)
print('Total no. of samples: Training and Testing dataset separately!')
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
```

```python
# Generate polynomial features of degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_train)
# Fit logistic regression model
model = LogisticRegression(solver='liblinear', max_iter=1000)
model.fit(X_poly, y_train)
```

```python
X_poly = poly.fit_transform(X_test)
```

```python
y_pred7 = model.predict(X_poly)
```

```python
y_pred7
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test, y_pred7) * 100
print(f"Logistic Regression model accuracy: {acc:.2f}%")
```
